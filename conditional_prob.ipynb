Conditional probability:

In Brife, conditional probability is the chances that some outcome (A) occurs given that another event (B) has also occurred. This is written as A given B, or as this formula: P(A|B), where the probability of A depends on that of B happening.

**Problem**:
Ten numbered cards are there from 1 to 15, and two cards a chosen at random such that *the sum of the numbers on both the cards is even*. Find the probability that the chosen cards are *odd-numbered*.

**solution**:
Let,
A ≡ event of selecting two odd-numbered cards
B ≡ event of selecting cards whose sum is even.

Then,
n(B) : number of ways of choosing two numbers whose sum is even (combination of two even number + combinations of two odd number).
n(B)  = 8C2 + 7C2.

n(A ∩ B) : number of ways of choosing odd-numbered cards such that their sum is even.
n(A ∩ B) = 8C2.

The probability of the choosen two cardes is odd given that the sum of them is even is,
P(A|B) = P(A ∩ B)/P(B) = n(A ∩ B)/n(B)
     
P(A|B)= 8C2 / (8C2 + 7C2) = 4/7.


```python
# Conditional probability problrm explained

# Importing important libraries
import math

# Function to calculate combination (n choose r)
def combinations(n, r):
    return math.factorial(n) / (math.factorial(r) * math.factorial(n - r))

# Function to calculate probability
def conditional_probability():
    # Number of ways of choosing two numbers whose sum is even
    n_B = combinations(8, 2) + combinations(7, 2)

    # Number of ways of choosing odd-numbered cards whose sum is even
    n_A_intersect_B = combinations(8, 2)

    # Probability of event A given B
    P_A_given_B = n_A_intersect_B / n_B

    return P_A_given_B

# Function to generate the event space of P(A|B)
def event_space():
    # Initialize an empty list to store the pairs of cards
    event_space_A_given_B = list()

    # Iterate through all possible pairs of cards
    for card1 in range(1, 16):
        for card2 in range(card1 + 1, 16):  # Ensure card2 is greater than card1
            # Check if both cards are odd-numbered and their sum is even
            if (card1 % 2 == 1) and (card2 % 2 == 1) and ((card1 + card2) % 2 == 0):
                # If conditions are met, add the pair (card1, card2) to the event space list
                event_space_A_given_B.append((card1, card2))

    # Return the event space list containing all valid pairs of cards
    return event_space_A_given_B

# Main function
def main():
    # Function call
    probability = conditional_probability()
    print("Probability that the chosen cards are odd-numbered given the sum is even:", probability)

    # Generate the event space of P(A|B)
    event_space_A_given_B = event_space()

    # Print the event space
    print("Event Space of P(A|B):", event_space_A_given_B)

if __name__ == "__main__":
    main()
```
